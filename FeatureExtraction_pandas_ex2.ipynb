```python
import pandas as pd
import numpy as np
```

```python
animeDf = pd.read_csv(r"sample_data/anime.csv")
```

```python
animeDf.head()
```

```python
def extract_eps(text):
  check = False
  data = ''
  for i in text:
    if i == ")":
      return data
    if i == "(":
      check = True
      continue
    if check:
      data+=i
```

```python
animeDf['Episodes'] = animeDf['Title'].apply(extract_eps)
```

```python
animeDf['Episodes'] = animeDf['Episodes'].str.replace(' eps','').astype(int)
```

```python
animeDf.loc[3]['Title']
```

```python
import re

def find_timestamp(text):
    pattern = r'[A-Z][a-z]{2}\s\d{4}\s*-\s*[A-Z][a-z]{2}\s\d{4}'
    match = re.search(pattern, text)
    return match.group(0) if match else None
```

```python
animeDf['Timestamp'] = animeDf['Title'].apply(find_timestamp)
animeDf.head(6)
```
