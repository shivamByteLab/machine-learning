Grouping

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
dt = pd.DataFrame(data)
```

```python
dt.groupby('Category')['Sales'].sum()
```

```python
dt.groupby('Store')['Sales'].sum()
```

```python
dt.groupby(['Store','Category'])['Sales'].sum()
```

Aggregation

```python
dt['Sales'].mode()
# """mean median min max count std """
```

```python
dt['Sales'].agg(['sum', 'mean', 'min', 'max', 'count', 'std', 'median'])
```

## PivotTables

```python


data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
               'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
            'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice']
}

pt = pd.DataFrame(data)

pt['Month'] = pt['Date'].dt.month_name()
pt['Quarter'] = 'Q' + pt['Date'].dt.quarter.astype(str)
pt

```

```python
pd.pivot_table(pt,values = "Sales",index = 'Region',columns="Product",aggfunc ='median')
```

```python
pivot2 = pd.pivot_table(pt, values=['Sales', 'Units'], index='Region', columns='Product')
pivot2
```

```python
pd.crosstab(pt['Region'],pt['Product'])
```
